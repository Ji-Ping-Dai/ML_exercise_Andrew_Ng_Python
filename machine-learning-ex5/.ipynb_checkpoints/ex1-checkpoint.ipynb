## 1. Linear regression

```python
#import modules
import numpy as np
import matplotlib.pyplot as plt 
import func
%matplotlib inline
plt.rc('text',usetex=True)
plt.rc('font',family='Times New Roman')
#read the data
data = np.loadtxt('data/ex1data1.txt',delimiter=',')
X = data[:,0].reshape([-1,1]) #input   #reshape!!!!!
y = data[:,1].reshape([-1,1]) #output
#plot the data
fig = plt.figure(figsize=[7,5])
ax = plt.axes()
func.plotdata(X,y,ax)
```

```python
# Add a column of ones to X (bias unit)
Xb = np.concatenate([np.ones([len(X),1]),X],axis=1)
# initialize fitting parameters
theta = np.zeros([X.shape[1]+1,1])
# Some gradient descent settings
iterations = 1500
alpha = 0.01
# Test cost function
J = func.costfunc(Xb,y,theta)
print('With theta = [0 ; 0],Cost computed = {0}'.format(J))
print('Expected cost value (approx) 32.07')
```

```python
# run gradient descent
theta_result, cost_iter = func.gradientDescent(Xb,y,theta,iterations,alpha)
print('Theta found by gradient descent is {0}'.format(theta_result.reshape(1,-1)))
print('Expected theta values (approx): [-3.6303, 1.1664]')
#plot the cost function with iterations
plt.plot(np.arange(1,iterations+1),cost_iter,color='g',ls='--')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.ylabel('Cost Function',size=15)
plt.xlabel('Iterations',size=15);
```

```python
#plot the linear fit
fig = plt.figure(figsize=[7,5])
ax = plt.axes()
func.plotdata(X,y,ax)
ax.plot(Xb[:,1],theta_result[0]+theta_result[1]*Xb[:,1],color='k',label='Linear regression')
ax.legend(fontsize=15)
```

```python
# Predict valuse
predict1 = np.array([1,7])@theta_result
print('For population = 35,000, we predict a profit of ${0:.2f}'.format(predict1[0]*10000))
```

```python
# Visualizing J(theta_0, theta_1)
theta_0 = np.linspace(-10,10,100)
theta_1 = np.linspace(-1,4,100)
T1, T2 = np.meshgrid(theta_0, theta_1)
cost2D = np.zeros([100,100])
for i in range(100):
    for j in range(100):
        Ta = np.array([[theta_0[i]],[theta_1[j]]])
        cost2D[i,j] = func.costfunc(Xb, y, Ta)
con=plt.contour(T1, T2, cost2D.T, np.logspace(-1,3,10)) #Because of the way meshgrids work
#an expample theta_0*theta1 T1*T2 
plt.clabel(con,inline=True)
plt.scatter(theta_0[31],theta_1[43],marker='*',c='r', s=100)
plt.xlabel(r'$\theta_0$',size=15)
plt.ylabel(r'$\theta_1$',size=15);
```

## 2. Linear regression with multiple variables

```python
#Load Data
data = np.loadtxt('data/ex1data2.txt', delimiter=',')
X = data[:,:2]
y = data[:,2].reshape([-1,1])
# Scale features and set them to zero mean
X, mu, sigma = func.featureNormalize(X)
# Add a column of ones to X (bias unit)
Xb = np.concatenate([np.ones([len(y),1]),X],axis=1)
```

```python
# same as one variable, because we use vectorization calculation
alpha = 0.1
iterations = 50
theta = np.zeros([X.shape[1]+1,1])
theta_result, cost_iter = func.gradientDescent(Xb,y,theta,iterations,alpha)
print('Theta found by gradient descent is {0}'.format(theta_result.reshape(1,-1)))
#plot the cost function with iterations
plt.plot(np.arange(1,iterations+1),cost_iter,color='g',ls='--')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.ylabel('Cost Function',size=15)
plt.xlabel('Iterations',size=15);
```

```python
# Estimate the price of a 1650 sq-ft, 3 br house
price=np.array([1, (1650-mu[0])/sigma[0], (3-mu[1])/sigma[1]])@theta_result
print('Predicted price of a 1650 sq-ft, 3 br house (using gradient descent) is ${0:.2f}'.format(price[0]))
```

```python
#Normal Equations
data = np.loadtxt('data/ex1data2.txt', delimiter=',')
X = data[:,:2]
y = data[:,2].reshape([-1,1])
Xb = np.concatenate([np.ones([len(y),1]),X],axis=1)
theta_result = np.linalg.inv(Xb.T@Xb)@Xb.T@y
print('Theta found by gradient descent is {0}'.format(theta_result.reshape(1,-1)))
price=np.array([1, 1650, 3])@theta_result
print('Predicted price of a 1650 sq-ft, 3 br house (using normal equation) is ${0:.2f}'.format(price[0]))
```
