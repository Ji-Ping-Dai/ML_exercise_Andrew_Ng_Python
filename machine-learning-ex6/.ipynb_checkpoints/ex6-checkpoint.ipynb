## 1. Support Vector Machines

```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.io import loadmat
import func
from sklearn.svm import SVC # support vector classifier
plt.rc('text',usetex=True)
plt.rc('font',family='Times New Roman')
%matplotlib inline
# Load data
dataMat = loadmat('data/ex6data1.mat')
X = dataMat['X']
y = dataMat['y']
fig = plt.figure(figsize=[7,5])
axes = plt.axes()
func.plotdata(X, y, axes)
axes.legend(['y=1','y=0'],fontsize=20);
```

```python
# Training Linear SVM
model = SVC(kernel='linear', C=1)
model.fit(X,y.ravel())
px = np.linspace(np.min(X[:,0]),np.max(X[:,0]),100)
py = np.linspace(np.min(X[:,1]),np.max(X[:,1]),100)
PX, PY =np.meshgrid(px, py)
xy = np.hstack([PX.reshape(-1,1),PY.reshape(-1,1)])
P = model.decision_function(xy).reshape(PX.shape)
fig = plt.figure(figsize=[7,5])
axes = plt.axes()
func.plotdata(X, y, axes) # for meshgrid plot, [0,0] is x0,y0, [0,1] is x1, y0....
# not like numpy [1,0] is x1, y0
axes.contour(PX,PY,P,colors='k',levels=[0],linestyles=['--']);
axes.plot([],[],color='k',ls='--')
axes.legend(['y=1','y=0', 'decision boundary'],fontsize=15);
```

```python
# Visualizing Dataset 2
dataMat = loadmat('data/ex6data2.mat')
X = dataMat['X']
y = dataMat['y']
model = SVC(kernel='rbf', C=1, gamma=1/2/0.1**2)
model.fit(X,y.ravel())
px = np.linspace(np.min(X[:,0]),np.max(X[:,0]),100)
py = np.linspace(np.min(X[:,1]),np.max(X[:,1]),100)
PX, PY =np.meshgrid(px, py)
xy = np.hstack([PX.reshape(-1,1),PY.reshape(-1,1)])
P = model.decision_function(xy).reshape(PX.shape)
fig = plt.figure(figsize=[7,5])
axes = plt.axes()
func.plotdata(X, y, axes)
axes.contour(PX,PY,P,colors='k',levels=[0],linestyles=['--']);
axes.plot([],[],color='k',ls='--')
axes.legend(['y=1','y=0', 'decision boundary'],fontsize=15);
```

```python
# Visualizing Dataset 3 and choose the optimal C and sigma
dataMat = loadmat('data/ex6data3.mat')
X = dataMat['X']
y = dataMat['y']
Xval = dataMat['Xval']
yval = dataMat['yval']
# find the optimal C and sigma
C, sigma = func.dataset3Params(X, y, Xval, yval)
# train and plot
model = SVC(kernel='rbf', C=C, gamma=1/2/sigma**2)
model.fit(X,y.ravel())
px = np.linspace(np.min(X[:,0]),np.max(X[:,0]),100)
py = np.linspace(np.min(X[:,1]),np.max(X[:,1]),100)
PX, PY =np.meshgrid(px, py)
xy = np.hstack([PX.reshape(-1,1),PY.reshape(-1,1)])
P = model.decision_function(xy).reshape(PX.shape) # decision_function return the Distance of the samples X to the separating hyperplane
# points between -1 and 1 are support vectors
fig = plt.figure(figsize=[7,5])
axes = plt.axes()
func.plotdata(X, y, axes)
axes.contour(PX,PY,P,colors='k',levels=[0],linestyles=['--']);
axes.plot([],[],color='k',ls='--')
axes.legend(['y=1','y=0', 'decision boundary'],fontsize=15);
```

## 2. Spam Classification with SVM

```python
# Extract Features
import pandas as pd
vocabList = pd.read_table('data/vocab.txt',header=None, names=['index'],index_col=1)
with open('data/emailSample1.txt') as f:
    file_contents = f.read()
word_indices = func.processEmail(file_contents)
features = np.zeros(vocabList.shape[0])
features[word_indices] = 1
```

```python
# Train Linear SVM for Spam Classification
dataMat = loadmat('data/spamTrain.mat')
X = dataMat['X']
y = dataMat['y']
C = 0.1
model = SVC(kernel='linear', C=C)
model.fit(X,y.ravel())
print('Training Accuracy: {0:.2f}%'.format(np.mean(model.predict(X)==y.ravel())*100))
```

```python
# Test Spam Classification
dataMat = loadmat('data/spamTest.mat')
Xtest = dataMat['Xtest']
ytest = dataMat['ytest']
print('Test Accuracy: {0:.2f}%'.format(np.mean(model.predict(Xtest)==ytest.ravel())*100))
```
