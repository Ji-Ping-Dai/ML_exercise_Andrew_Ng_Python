## 1. Linear Regression

```python
#import modules
import numpy as np
import matplotlib.pyplot as plt 
import func
%matplotlib inline
plt.rc('text',usetex=True)
plt.rc('font',family='Times New Roman')
#read the data
data = np.loadtxt('data/ex1data1.txt',delimiter=',')
X = data[:,0].reshape([-1,1]) #input
y = data[:,1].reshape([-1,1]) #output
#plot the data
func.plotdata(X,y)
```

```python
# Add a column of ones to X (bias unit)
Xb = np.concatenate([np.ones([len(X),1]),X],axis=1)
# initialize fitting parameters
theta = np.zeros([X.shape[1]+1,1])
# Some gradient descent settings
iterations = 1500
alpha = 0.01
# Test cost function
J = func.costfunc(Xb,y,theta)
print('With theta = [0 ; 0],Cost computed = {0}'.format(J))
print('Expected cost value (approx) 32.07')
```

```python
# run gradient descent
theta_result, cost_iter = func.gradientDescent(Xb,y,theta,iterations,alpha)
```
