## 1. Logistic Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import func
%matplotlib inline
plt.rc('text',usetex=True)
plt.rc('font',family='Times New Roman')
#Load Data
data = np.loadtxt('data/ex2data1.txt',delimiter=',')
X = data[:,:2]
y = data[:,2].reshape(-1,1)
#plotting
fig = plt.figure(figsize=[7,5])
ax = plt.axes()
func.plotdata(X,y,ax)
ax.set_xlabel('Exam 1 score',size=20)
ax.set_ylabel('Exam 2 score',size=20)
plt.legend(['Admitted','Not Admitted'],fontsize=15);
```

```python
# Add a column of ones to X (bias unit)
Xb = np.concatenate([np.ones([len(X),1]),X],axis=1)
# initialize fitting parameters
theta = np.zeros(X.shape[1]+1)
#Compute and display initial cost and gradient
cost = func.costfunc(theta,Xb,y) # theta must be the first arg, due to the properties of minimize
grad = func.grid(theta,Xb,y)
print('Expected cost at initial theta (zeros) is {0}'.format(0.693))
print('Cost at initial theta (zeros) is {0}'.format(cost[0,0]))
print('Expected Gradient at initial theta (zeros) is {0}'.format([-0.1000,-12.0092,-11.2628]))
print('Gradient at initial theta (zeros) is {0}'.format(grad))
# Run minimize to obtain the optimal theta, due to the properties of minimize, we need theta to be 1D, jac to be 1D
from scipy.optimize import minimize
result=minimize(fun=func.costfunc, x0=theta, method='BFGS', jac=func.grid, args=(Xb,y))
print('Cost at theta found by minimize is {0}'.format(result.fun))
print('theta are {0}'.format(result.x))
theta = result.x
# Try gradient descent #toooooooooooooooooooo slowwwwwwwwwwwwwwwwwwwwwwwwwww
#alpha = 0.001
#iterations = 200000
#theta_GD, cost_GD = func.gradientDescent(Xb, y, np.zeros([X.shape[1]+1,1]), iterations, alpha)
```

```python
# Plot Boundary
fig = plt.figure(figsize=[7,5])
ax = plt.axes()
func.plotboundary(X,y,theta,ax)
ax.set_xlabel('Exam 1 score',size=20)
ax.set_ylabel('Exam 2 score',size=20)
plt.legend(['Admitted','Not Admitted','Decision Boundary'],fontsize=15);
```

```python
#Predict and Accuracies
p = func.predict(Xb,theta)
print('The Train Accuracy is: {0:.0f}%'.format(np.mean(p==y)*100))
print('Expected accuracy (approx): 89%')
```

## 2. Regularized Logistic Regression


```python
import numpy as np
import matplotlib.pyplot as plt
import func
%matplotlib inline
plt.rc('text',usetex=True)
plt.rc('font',family='Times New Roman')
# load data
data = np.loadtxt('data/ex2data2.txt', delimiter=',')
X = data[:,:2]
y = data[:,2].reshape(-1,1)
#plotting
fig = plt.figure(figsize=[7,5])
ax = plt.axes()
func.plotdata(X,y,ax)
ax.set_xlabel('Microchip Test 1',size=20)
ax.set_ylabel('Microchip Test 2',size=20)
plt.legend(['y=1','y=0'],fontsize=15);
```

```python
# adds columns for higher order
Xb = func.mapX(X[:,0],X[:,1])
initial_theta = np.zeros(Xb.shape[1])
lam = 1
#Compute and display initial cost and gradient
cost = func.costfuncReg(initial_theta,Xb,y,lam) # theta must be the first arg, due to the properties of minimize
grad = func.gridReg(initial_theta,Xb,y,lam)
print('Expected cost at initial theta (zeros) is {0}'.format(0.693))
print('Cost at initial theta (zeros) is {0}'.format(cost[0,0]))
```

```python
# Run minimize to obtain the optimal theta, due to the properties of minimize, we need theta to be 1D, jac to be 1D
from scipy.optimize import minimize
result=minimize(fun=func.costfuncReg, x0=initial_theta, method='BFGS', jac=func.gridReg, args=(Xb,y,lam))
print('Cost at theta found by minimize is {0}'.format(result.fun))
print('theta are {0}'.format(result.x))
theta = result.x
```

```python
# Plot Boundary
fig = plt.figure(figsize=[7,5])
ax = plt.axes()
func.plotboundary(X,y,theta,ax)
ax.set_xlabel('Microchip Test 1',size=20)
ax.set_ylabel('Microchip Test 2',size=20)
ax.plot([],[],ls='-',c='k')
plt.legend(['y=1','y=0','Decision Boundary'],fontsize=15);
```

```python
# Compute accuracy on our training set
p = func.predict(Xb,theta)
print('The Train Accuracy is: {0:.1f}%'.format(np.mean(p==y)*100))
print('Expected accuracy (approx): 83.1%')
```
