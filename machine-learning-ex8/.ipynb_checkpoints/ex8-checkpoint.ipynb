## 1. Anomaly Detection

```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.io import loadmat
import func
dataMat = loadmat('data/ex8data1.mat')
# Load an example dataset that we will be using
X = dataMat['X']
Xval = dataMat['Xval']
yval = dataMat['yval']
```

```python
# Estimate the dataset statistics
mu, sigma2 = func.estimateGaussian(X)
p = func.multivariateGaussian(X, mu, sigma2)
# Visualize the date set
fig = plt.figure()
axes = plt.axes()
func.visualizeFit(X, mu, sigma2,axes);
# Find Outliers
pval = func.multivariateGaussian(Xval, mu, sigma2)
epsilon, F1 = func.selectThreshold(yval,pval)
outliers = np.where(p<epsilon)
axes.scatter(X[outliers,0], X[outliers, 1],color="w",linewidths=1,s=100,edgecolors='r');
```

```python
# Multidimensional Outliers
dataMat = loadmat('data/ex8data2.mat')
X = dataMat['X']
Xval = dataMat['Xval']
yval = dataMat['yval']
mu, sigma2 = func.estimateGaussian(X)
p = func.multivariateGaussian(X, mu, sigma2)
pval = func.multivariateGaussian(Xval, mu, sigma2)
epsilon, F1 = func.selectThreshold(yval,pval)
print('Outliers found: {0}'.format(sum(p<epsilon)))
```

## 2. Collaborative Filtering

```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.io import loadmat
import func
# Loading movie ratings dataset
dataMat = loadmat('data/ex8_movies.mat')
Y = dataMat['Y']
R = dataMat['R']
print('Average rating for movie 1 (Toy Story): {0:.1f}/5'.format(np.mean(Y[0,R[0,:]>0])))
```

```python
# Collaborative Filtering Cost Function
dataMat = loadmat('data/ex8_movieParams.mat')
X = dataMat['X']
Theta = dataMat['Theta']
# Reduce the data set size so that this runs faster
num_users = 4
num_movies = 5
num_features = 3
X = X[:num_movies,:num_features]
Theta = Theta[:num_users,:num_features]
Y = Y[:num_movies,:num_users]
R = R[:num_movies,:num_users]
# Evaluate cost function and gradident
J = func.cofiCostFunc(np.concatenate([X.ravel(),Theta.ravel()]),\
                      Y,R,num_users,num_movies,num_features,1.5)
grad = func.cofigrad(np.concatenate([X.ravel(),Theta.ravel()]),\
                      Y,R,num_users,num_movies,num_features,1.5)
# numerical estimate
P = np.concatenate([X.ravel(),Theta.ravel()])
P_pert = np.zeros(len(P))
diff = np.zeros(len(P))
for i in range(len(P_pert)):
    P_pert[i] = 1e-4
    J_pertp = func.cofiCostFunc(P+P_pert,\
                      Y,R,num_users,num_movies,num_features,1.5)
    J_pertm = func.cofiCostFunc(P-P_pert,\
                      Y,R,num_users,num_movies,num_features,1.5)
    diff[i] = (J_pertp-J_pertm)/(2e-4)
    P_pert[i] = 0
print('Relative difference between numerical Ggadient and analytical gradient:\n {0}'.format(np.mean(grad-diff)/np.mean(grad)))
```

```python
# Learning Movie Ratings
dataMat = loadmat('data/ex8_movies.mat')
Y = dataMat['Y']
Y=Y.astype(np.float64)
R = dataMat['R']
for i in range(Y.shape[0]):
    Y[i,R[i,:]>0] = Y[i,R[i,:]>0] - np.mean(Y[i,R[i,:]>0])
num_users = Y.shape[1]
num_movies = Y.shape[0]
num_features = 10
# Set Initial Parameters (Theta, X)
X = np.random.randn(num_movies, num_features)
Theta = np.random.randn(num_users, num_features)
initial_parameters = np.concatenate([X.ravel(),Theta.ravel()])
lam = 10
from scipy.optimize import minimize
#minimize, It provides an alternative way to call ``fmin_cg``, by specifying ``method='CG'``.
args = (Y,R,num_users,num_movies,num_features,lam)
result = minimize(fun=func.cofiCostFunc, x0=initial_parameters, method='CG', \
                  jac=func.cofigrad,args=args, options={'maxiter':50})
```

```python
# Recommendation
params = result.x
X = params[:num_movies*num_features].reshape(num_movies,num_features)
Theta = params[num_movies*num_features:].reshape(num_users, num_features)
p = X@Theta.T
dataMat = loadmat('data/ex8_movies.mat')
Y = dataMat['Y']
Y=Y.astype(np.float64)
R = dataMat['R']
for i in range(p.shape[0]):
    p[i,R[i,:]==0] = p[i,R[i,:]==0] + np.mean(Y[i,R[i,:]>0])
    p[i,R[i,:]>0] = Y[i,R[i,:]>0]
```

```python
print('Average rating for movie 1 (Toy Story): {0:.1f}/5'.format(np.mean(p[0,R[0,:]>0])))
```
