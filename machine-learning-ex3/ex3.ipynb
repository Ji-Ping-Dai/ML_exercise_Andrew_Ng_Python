## 1. Logistic Regression  One-vs-all

```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.io import loadmat
import func
plt.rc('text',usetex=True)
plt.rc('font',family='Times New Roman')
%matplotlib inline
#load data
dataMat = loadmat('data/ex3data1.mat')
X = dataMat['X']
y = dataMat['y']
#print(X.shape)
# show some of them
fig, ax = plt.subplots(10,10,figsize=(8,8))
for i in range(10):
    for j in range(10):
        n = np.random.randint(0,5000)
        ax[i,j].imshow(X[n,:].reshape(20,20,order='F'))
        ax[i,j].set(xticks=[],yticks=[])
```

```python
#One-vs-All Training
lam = 0.1
labels = len(np.unique(y))
Xb = np.concatenate([np.ones([len(X),1]),X],axis=1)
all_theta = func.onevsall(Xb, y, labels, lam)
```

```python
#Predict for One-Vs-All
pred = func.predictonevsall(all_theta,Xb)
print('The Train Accuracy is about {0:.1f}%'.format(np.mean((pred == y)) * 100))
# Randomly permute examples
n = np.random.randint(0,5000)
plt.imshow(X[n,:].reshape(20,20,order='F'))
plt.axis('off');
p = func.predictonevsall(all_theta, Xb[n,:].reshape(1,-1))
print('Neural Network Prediction: {0}'.format(p[0,0]%10))
```

## 2. Neural Networks

```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.io import loadmat
import func
plt.rc('text',usetex=True)
plt.rc('font',family='Times New Roman')
# Setup the parameters you will use for this exercise
input_layer_size = 400
gidden_layer_siez = 25
num_labels = 10
# Loading data
dataMat = loadmat('data/ex3data1.mat')
X = dataMat['X']
y = dataMat['y']
Xb = np.concatenate([np.ones([len(X),1]),X],axis=1)
# Loading parameters (weight)
parMat = loadmat('data/ex3weights.mat')
Theta1 = parMat['Theta1']
Theta2 = parMat['Theta2']
```

```python
# Implement Predict
pred = func.predictnn(Theta1, Theta2, Xb)
print('The Train Accuracy is about {0:.1f}%'.format(np.mean((pred == y)) * 100))
```

```python
# Randomly permute examples
n = np.random.randint(0,5000)
plt.imshow(X[n,:].reshape(20,20,order='F'))
plt.axis('off');
p = func.predictnn(Theta1, Theta2, Xb[n,:].reshape(1,-1))
print('Neural Network Prediction: {0}'.format(p[0,0]%10))
```
